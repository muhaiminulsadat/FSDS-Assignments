### Problem 1: Inventory Price Calculator

```python
def calculate_inventory_summary(prices):
    total_cost = sum(prices)
    average_cost = total_cost / len(prices)
    return total_cost, average_cost

prices = [120, 250, 399, 150]
print(calculate_inventory_summary(prices))
```

### Problem 2: Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    unique_visitors = sorted(set(visitor_list))
    length = len(unique_visitors)

    unique_str = ", ".join(str(x) for x in unique_visitors)


    return (f"{length} (The unique IDs are {unique_str})")

count_unique_visitors([101, 205, 101, 310, 205, 550])
```

### Problem 3: Product Category Counter

```python
def count_categories(products):

    category_count = {}

    for product in products:
        
        if product in category_count:
            category_count[product] += 1
        else:
            category_count[product] = 1
    
    return category_count

print(count_categories(["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]))
```

### Problem 4: Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high = []
    normal = []

    for temp in temp_list:
        if temp> 38:
            high.append(temp)
        else: 
            normal.append(temp)
    return high,normal

temperatures = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

high, normal = analyze_temperatures(temperatures)

print(f"High List (>38째C): {high}")
print(f"Normal List (<=38째C): {normal}")
print(f"Full Output Tuple: {analyze_temperatures(temperatures)}")
```

### Problem 5: Menu Price Lookup

```python
def get_price(menu: dict, item):
    if item in menu:           
        return menu[item]
    else:
        return "Item not found"

menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
print(get_price(menu, "Muffin"))  
print(get_price(menu, "Donut")) 
```

### Problem 6: Student Grading System

```python
def calculate_grade(score):
    if score >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 60:
        return "A-"
    else:
        return "F"

def grade_students(student_dict):
    grades_dict = {}
    for student, score in student_dict.items():
        grades_dict[student] = calculate_grade(score)
    return grades_dict

scores = {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}

print(grade_students(scores))
```

### Problem 7: Discount Calculator

```python
def apply_discounts(products):
    updated_prices = []
    for name, price, discount in products:
        discounted_price = price * (1 - discount / 100) 
        updated_prices.append(round(discounted_price, 2))
    return updated_prices

products = [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]

print(apply_discounts(products))
```

### Problem 8: Hashtag Generator

```python
def generate_hashtags(sentence):
    words = sentence.split()
    hashtags = ["#" + word.lower() for word in words]
    return hashtags

print(generate_hashtags("AI Data Science"))
```

### Problem 10: Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
    total = 0
    for item in cart:
        if item in prices:         
            total += prices[item]  
    return round(total, 2)        

cart = ["Milk", "Bread", "Milk", "Cheese", "Bread"]
prices = {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

print(calculate_total(cart, prices))
```
